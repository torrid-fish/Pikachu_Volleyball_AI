```python
import numpy as np
import matplotlib.pyplot as plt
import os
from IPython.display import clear_output
while True:
    clear_output(wait=True)
    winrt = np.load('history_winrt.npy')
    loss = np.load('history_loss.npy')
    ax1 = plt.subplot(1, 2, 1)
    ax1.plot(winrt)
    ax1.legend([f'A{i}' for i in range(winrt.shape[1])])
    ax1.set_title('Previous Win Rate')
    ax2 = plt.subplot(1, 2, 2)
    ax2.plot(loss)
    ax2.legend([f'A{i}' for i in range(winrt.shape[1])])
    ax2.set_ylim(0, 0.1)
    ax2.set_title('Loss')
    plt.tight_layout(pad= 0.5)
    plt.pause(5)
    plt.show()
```
